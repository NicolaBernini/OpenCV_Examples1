```
import cv2
import numpy as np 
from matplotlib import pyplot as plt 
import urllib.request
```

```

def bs2image(s): 
  #np_array = np.asarray(repr, dtype="uint8")
  np_array = np.frombuffer(s, np.uint8)
  return cv2.imdecode( np_array, cv2.IMREAD_COLOR )

# Defines an API to define the "URL to File Path" Mol 
def url2img(url): 
  # NOTE: Define a "Local Endpoint PP" Mol which provides the "Read Interface Path" Mol 
  with urllib.request.urlopen(url) as local_ep:
    bs = local_ep.read()
    return bs2image(bs)

  
```

```
img_left = "https://www.dropbox.com/s/6b4zgvzentaa99c/921919841_a30df938f2_o.jpg?raw=1"
img_right = "https://www.dropbox.com/s/cdsttf9vp7lwarv/4191453057_c86028ce1f_o.jpg?raw=1"

im_left = url2img(img_left)
im_right = url2img(img_right)
```
